Data Reduction

```python
df = df.drop(['S.No.'], axis = 1)
df.info()
```

 Creating Features

Adding a new column to understand what is the age of the Car.

```python
from datetime import date
df['Car_Age'] =  date.today().year-df['Year']
```

```python
df.head()
```

Let’s split the name and introduce new variables “Brand” and “Model”

```python
#df.Name.str.split().str.get(0)

df['Brand'] = df.Name.str.split().str.get(0)
```

```python
df.head()
```

```python
# Lets get the Model
df['Model'] = df.Name.str.split().str.get(1) + df.Name.str.split().str.get(2)
```

```python
df.head()
```

```python
# Lets have a look at the new columns that we have added
df[['Name','Brand','Model']]
```

 Data Cleaning/Wrangling

```python
print(df.Brand.unique())
print(df.Brand.nunique())
```

```python
searchfor = ['Isuzu' ,'ISUZU','Mini','Land']
df[df.Brand.str.contains('|'.join(searchfor))].head(5)
```

```python
# Replacig the correct names 

df["Brand"].replace({"ISUZU": "Isuzu", "Mini": "Mini Cooper","Land":"Land Rover"}, inplace=True)
```

Before we do EDA, lets separate Numerical and categorical variables for easy analysis

```python
cat_cols=df.select_dtypes(include=['object']).columns
num_cols = df.select_dtypes(include=np.number).columns.tolist()
print("Categorical Variables:")
print(cat_cols)
print("Numerical Variables:")
print(num_cols)
```

categorical variables are being visualized using a count plot. Categorical variables provide the pattern of factors influencing car price

```python
# Lets understand what kind of Categorical variables we have and how they have distributed.

# We have below Categorical variavles  
# ['Name', 'Location', 'Fuel_Type', 'Transmission', 'Owner_Type','Mileage', 'Engine', 'Power', 'New_Price', 'Brand', 'Model']


df['Fuel_Type'].value_counts()
#df['Location'].value_counts()
```
