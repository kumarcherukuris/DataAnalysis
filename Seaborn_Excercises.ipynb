```python
# import libs
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load the data
df = pd.read_csv('C:\\Users\\user\\Workshop_Files\\DataSets\\tips.csv')
```

```python
df.head()
```

```python
# Get the tip based on Day

sns.lineplot(x="day" , y="tip", data = df)
plt.show()
```

```python
# Plot a chart that gives Total bill per day by Male/Female

# properties
# palette

sns.lineplot(x="day" , y="total_bill", data = df, hue='sex', palette='flare')
plt.show()
```

```python
# Create a bar plot for the same.
sns.barplot(x="day", y="total_bill", data= df)
plt.show()

# Get by male / febale 
#  hue="sex"
```
